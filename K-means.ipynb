```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("income.csv")
```

```python
df.head()
```

```python
plt.scatter(df["Age"],df["Income($)"])
```

```python
km=KMeans(n_clusters=3)
```

```python
df["cluster"]=km.fit_predict(df[["Age","Income($)"]])
```

```python
c1=df[df.cluster==0]
c2=df[df.cluster==1]
c3=df[df.cluster==2]
```

```python
c1
```

```python
plt.scatter(c1.Age,c1["Income($)"],color="Black")
plt.scatter(c2.Age,c2["Income($)"],color="Blue")
plt.scatter(c3.Age,c3["Income($)"],color="Red")
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar=MinMaxScaler()
scalar.fit(df[["Income($)"]])
df["Income($)"]=scalar.transform(df[["Income($)"]])
scalar.fit(df[["Age"]])
df["Age"]=scalar.transform(df[["Age"]])
df
```

```python
df["cluster"]=km.fit_predict(df[["Age","Income($)"]])
centers=km.cluster_centers_
centers
```

```python
centers[:,0]
```

```python
centers[:,1]
```

```python

c1=df[df.cluster==0]
c2=df[df.cluster==1]
c3=df[df.cluster==2]
plt.scatter(centers[:,0],centers[:,1],marker="*")
plt.scatter(c1.Age,c1["Income($)"],color="Black")
plt.scatter(c2.Age,c2["Income($)"],color="Blue")
plt.scatter(c3.Age,c3["Income($)"],color="Red")
```
