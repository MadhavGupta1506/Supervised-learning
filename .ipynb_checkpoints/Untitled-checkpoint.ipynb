```python
res_id=pd.DataFrame(["" for _ in range(len(data))])
```

```python
data['res_id']=res_id
```

```python
for i in range(len(data)):
    for j in range(len(indr)):
        if data.at[i, 'Restaurant'] == indr.at[j, 'restaurant_name']:
            data.at[i, 'res_id'] = indr.at[j, 'res_id']
            print(data.at[i, 'res_id'], indr.at[j, 'res_id'])
            break  # break inner loop once matched
```
