```python
import pandas as pd
# from sklearn.preprocessing import 
```

```python
df=pd.read_csv("titanic.csv")
df.drop(df[["PassengerId","Cabin","Pclass","Name","SibSp","Parch","Ticket",'Embarked']],axis='columns',inplace=True)
```

```python
target=df.Survived
inputs=df.drop(['Survived'],axis='columns')
```

```python
dummies=pd.get_dummies(df.Sex,dtype=int)
```

```python
inputs=pd.concat([inputs,dummies],axis="columns",)
```

```python
inputs.drop(["Sex"],axis="columns",inplace=True)
```

```python
inputs.Age.fillna(inputs['Age'].median(),inplace=True)
```

```python
inputs.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=.3)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
